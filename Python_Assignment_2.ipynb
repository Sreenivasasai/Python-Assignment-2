###  Part 1 : Beginner Level

```python
### Simple For Loop
## Write a Python program that takes an integer input N and uses a for loop to print all numbers from 1 up to and including N.

N = int(input("Enter any Number(N):"))
print(f"The numbers from 1 to {N} are:")
for i in range(1,N+1):
  print(i)
```

```python
### Simple While Loop
## Write a program that uses a while loop to print the numbers from 10 down to 1.

i=10
while i>=1:
  print(i)
  i-=1
```

```python
### Basic Conditional Check
## Ask the user for their age.Use a single if/else statement to print "you are an adult." if the age is 18 0r greater, and "you are a minor." otherwise.

age = int(input("Enter your age:"))
if age >= 18:
  print("you are an adult.")
else:
  print("you are a minor")
```

```python
### Looping and Filtering
## Write a program that iterates through the list ['apple', 'banana', 'cherry', 'grape', 'blackberry'] and prints only the fruits that start with the letter 'b'.

fruits = ['apple','banana','cherry','grape','blackberry']
for fruit in fruits:
  if fruit.startswith('b'):
    print(fruit)
```

```python
### Grade Classification
## Take a numerical grade (0-100) as input. Use if/elif/else to print the corresponding letter grade:
## * "A" for 90-100
## * "B" for 80-89
## * "C" for 70-79
## * "F" for below 70

grade = int(input("Enter your grade(0-100):"))
if grade >= 90:
  print("A")
elif grade >=80:
  print("B")
elif grade >= 70:
  print("C")
else:
  print("F")
```

```python
### Input Until Done
## Write a program that uses a while loop to continuously ask the user for a series of numbers,stopping only when the user enters the word "done".After they enter "done",print the largest number they entered.

largest = None
while True:
  value = input("Enter a number :")
  if value == "done":
    break
  num = int(value)
  if largest is None or num > largest:
    largest = num
print("Largest number:",largest)
```

```python
### Vowel Counter
## Use a for loop and conditionals (if) to count and print the total number of vowels (a, e, i, o, u) in the string: programmingisfun.

text = "programmingisfun"
vowels = "aeiou"
count = 0
for ch in text :
  if ch in vowels:
    count +=1
print(count)
```

```python
### Factorial Calculation
## Write a program to find the factorial of a user-input number N (e.g., 5! = 5 \times 4 \times 3 \times 2 \times 1). Use a for loop.

n = int(input("Enter a number:"))
fact = 1
for i in range(1,n+1):
  fact=fact*i
print("factorial is:",fact)
```

```python
### Basic Pattern Printing
## Write a program that takes an integer N and prints a right-angled triangle pattern of * characters with N rows.

n = int(input("Enter a number:"))
for i in range(1,n+1):
  print("*" * i)
```

```python
### Complex Filtering with Length
## Write a program that iterates through a list of names ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Irene']. If a name is longer than 4 characters AND starts with a vowel ('A', 'E', 'I', 'O', 'U'), print the name.

names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Irene']
vowels = ('A','E','I','O','U')
for name in names:
  if len(name) > 4 and name.startswith(vowels):
    print(name)
```
